```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
raw_df = pd.read_csv('Data\officialbhafc.csv')
```

```python
raw_df
```

Tách riêng để phân loại các reactions

```python
raw_df['reactions'] = raw_df['reactions'].apply(lambda x : dict(eval(x)) )
df_detailReactions = raw_df['reactions'].apply(pd.Series )
     
```

```python
raw_df = pd.concat([raw_df, df_detailReactions], axis=1).drop('reactions',axis=1)
raw_df[['post_id','thích','yêu thích','haha','wow','buồn','phẫn nộ','thương thương',
                                 'shares','comments','reaction_count']]
```

Loại bỏ những hàng mà không có giá trị của những trường chính

```python
raw_df=raw_df.dropna(subset=['text','post_text','reaction_count','video'],how='all')
```

```python
raw_df=raw_df.dropna(subset=['reaction_count'],how='all')
```

Xóa đi các cột và các hàng không chứa giá trị

```python
raw_df=raw_df.dropna(how='all')
```

```python
raw_df=raw_df.dropna(axis=1,how='all')
```

Kiểm tra và loại bỏ những dòng dữ liệu trùng lặp trong dataframe

```python
raw_df=raw_df.drop_duplicates()
```

```python
raw_df.fillna(0,inplace=True)
```

```python
raw_df.info()
```

```python
raw_df
```

Loại bỏ emoji, các ký tự đặc biệt và url

```python
import re

def clean_text(text):

    if isinstance(text,str):
        # loại bỏ các emoji
        text = re.sub(r'[\U00010000-\U0010ffff]','', text)

        # loại bỏ các ký tự đặc biệt 
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

        # loại bỏ các đường dẫn
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

        return text
raw_df['post_text']=raw_df['post_text'].apply(clean_text)
```

Chuyển kiểu dữ liệu từ fetched time sang datetime và tách cột datetime thành có trường nhỏ hơn

```python
# Chuyển kiểu dữ liệu của fetched_time sang datetime64
raw_df['time'] = pd.to_datetime(raw_df['time'])

raw_df['hour'] = raw_df['time'].dt.hour
raw_df['day'] = raw_df['time'].dt.day
raw_df['month'] = raw_df['time'].dt.month
raw_df['year'] = raw_df['time'].dt.year
raw_df[['hour', 'day', 'month', 'year']]
```
