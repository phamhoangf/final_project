Lo·∫°i b·ªè emoji, c√°c k√Ω t·ª± ƒë·∫∑c bi·ªát v√† url

```python
import re

def clean_text(text):

    if isinstance(text,str):
        # lo·∫°i b·ªè c√°c emoji
        text = re.sub(r'[\U00010000-\U0010ffff]','', text)

        # lo·∫°i b·ªè c√°c k√Ω t·ª± ƒë·∫∑c bi·ªát 
        text = re.sub(r'[^a-zA-Z0-9\s]', '', text)

        # lo·∫°i b·ªè c√°c ƒë∆∞·ªùng d·∫´n
        text = re.sub(r'http\S+|www\S+|https\S+', '', text, flags=re.MULTILINE)

        return text
raw_df['post_text']=raw_df['post_text'].apply(clean_text)
```

Chuy·ªÉn ki·ªÉu d·ªØ li·ªáu t·ª´ fetched time sang datetime v√† t√°ch c·ªôt datetime th√†nh c√≥ tr∆∞·ªùng nh·ªè h∆°n

```python
# Chuy·ªÉn ki·ªÉu d·ªØ li·ªáu c·ªßa fetched_time sang datetime64
raw_df['time'] = pd.to_datetime(raw_df['time'])

raw_df['hour'] = raw_df['time'].dt.hour
raw_df['day'] = raw_df['time'].dt.day
raw_df['month'] = raw_df['time'].dt.month
raw_df['year'] = raw_df['time'].dt.year
raw_df[['hour', 'day', 'month', 'year']]
```
